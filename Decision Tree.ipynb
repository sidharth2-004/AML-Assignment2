```python
from chefboost import Chefboost as chef
import pandas as pd
```

```python
df = pd.read_csv("liver_cirrhosis.csv")
df.head()
```

```python
df=df.iloc[0:50,:]
```

```python
config = {'algorithm': 'ID3'}
model = chef.fit(df, config = config, target_label = 'Stage')
```
